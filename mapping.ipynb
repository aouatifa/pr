```python
from datetime import timedelta

def convertir_chaine_en_duree(chaine_duree):
    if '.' in chaine_duree:
      parties = [int(partie) for partie in chaine_duree.split('.')]
    else:
      # Diviser la cha√Æne en parties en utilisant le d√©limiteur ":"
      parties = [int(partie) for partie in chaine_duree.split(':')]

    # Ajouter des parties manquantes (heures, minutes, secondes)
    while len(parties) < 3:
        parties.insert(0, 0)

    # Retourner un objet timedelta
    return timedelta(hours=parties[0], minutes=parties[1], seconds=parties[2])

def formater_duree(chaine_duree):
    # Convertir la cha√Æne en objet timedelta
    duree = convertir_chaine_en_duree(chaine_duree)

    # Formater le r√©sultat en cha√Æne de dur√©e (HH:MM:SS)
    heures, reste = divmod(duree.seconds, 3600)
    minutes, secondes = divmod(reste, 60)

    # Retourner la cha√Æne format√©e
    return "{:02}:{:02}:{:02}".format(heures, minutes, secondes)
```

```python
def normalize_title(title):
    return title.lower().strip()

# Exemple pour Facebook
transformed_data_facebook = []
for video in data_facebook:
    normalized_title = normalize_title(video['Title'])
    duration = formater_duree(video['Duration'])  # Assurez-vous que la dur√©e est normalis√©e
    transformed_data_facebook.append({
        'titre': normalized_title,
        'duree': duration,
        'plateforme': 'Facebook'
    })
```
