```python
post_infos = []

for i in range(3):
    post_info = {}
    
    print("POST NUM ", i, " : ")
    
    view_count = views[i].text
    print("NEMBER OF VIEWS : ", view_count)
    post_info['NEMBER OF VIEWS'] = view_count
    
    elem_hover = driver.find_element(By.XPATH, '(//div[@class="_aajz"])[' + str(i + 1) + ']')
    actions.move_to_element(elem_hover)
    driver.implicitly_wait(60)
    actions.perform()
    
    ul = driver.find_element(By.XPATH, '(//ul[@class="_abpo"])[' + str(i + 1) + ']')
    li = ul.find_elements(By.TAG_NAME, 'li')
    like_count = li[0].text
    print("NEMBER OF LIKES : " + like_count)
    post_info['NEMBER OF LIKES'] = like_count
    
    comment_count = li[1].text
    print("NEMBER OF COMMENTS : " + comment_count)
    post_info['NEMBER OF COMMENTS'] = comment_count
    
    actions.click(elem_hover)
    driver.implicitly_wait(60)
    actions.perform()
    
    title = driver.find_element(By.XPATH, '(//span[@class="_aacl _aaco _aacu _aacx _aad7 _aade"])[1]')
    title_text = title.text.split('\n', 1)[0]
    print("Titre ", title_text)
    post_info['Titre'] = title_text
    
    date = driver.find_element(By.TAG_NAME, 'time').get_attribute('datetime')
    print("Date de publication ", date)
    post_info['Date de publication'] = date
    
    subcomments = driver.find_elements(By.XPATH, ' //span[@class="_a9yi"]')
    subcomment_texts = [subcomment.text for subcomment in subcomments]
    
    comments = driver.find_elements(By.XPATH, '//ul[@class="_a9ym"]//span[@class="_aacl _aaco _aacu _aacx _aad7 _aade"]')
    comment_texts = [comment.text for comment in comments]
    
    post_info['Subcomments'] = subcomment_texts
    post_info['Comments'] = comment_texts
    
    post_infos.append(post_info)
    
    actions.perform()

# Imprimez la liste des informations
print(post_infos)
```

```python
import json

# CrÃ©ez une liste pour stocker les informations des posts
post_infos = []

for i in range(3):
    post_info = {}
    
    print("POST NUM ", i, " : ")
    print("NEMBER OF VIEWS : ", views[i].text)
    post_info["NEMBER OF VIEWS"] = views[i].text

    elem_hover = driver.find_element(By.XPATH, '(//div[@class="_aajz"])[' + str(i + 1) + ']')
    actions.move_to_element(elem_hover)
    driver.implicitly_wait(60)
    actions.perform()

    ul = driver.find_element(By.XPATH, '(//ul[@class="_abpo"])[' + str(i + 1) + ']')
    li = ul.find_elements(By.TAG_NAME, 'li')
    print("NEMBER OF LIKES : " + li[0].text)
    print("NEMBER OF COMMENTS : " + li[1].text)
    post_info["NEMBER OF LIKES"] = li[0].text
    post_info["NEMBER OF COMMENTS"] = li[1].text

    actions.click(elem_hover)
    driver.implicitly_wait(60)
    actions.perform()
    title = driver.find_element(By.XPATH, '(//span[@class="_aacl _aaco _aacu _aacx _aad7 _aade"])[1]')
    title_text = title.text.split('\n', 1)[0]
    print("Titre ", title_text)
    post_info["Titre"] = title_text

    date = driver.find_element(By.TAG_NAME, 'time').get_attribute('datetime')
    print("Date de publication ", date)
    post_info["Date de publication"] = date

    subcomments = driver.find_elements(By.XPATH, ' //span[@class="_a9yi"]')
    subcomment_texts = [subcomment.text for subcomment in subcomments]
    post_info["Subcomments"] = subcomment_texts

    comments = driver.find_elements(By.XPATH, '//ul[@class="_a9ym"]//span[@class="_aacl _aaco _aacu _aacx _aad7 _aade"]')
    comment_texts = [comment.text for comment in comments]
    post_info["Comments"] = comment_texts

    post_infos.append(post_info)

# Convertissez la liste de dictionnaires en format JSON
json_data = json.dumps(post_infos, indent=4)

# Imprimez ou enregistrez les donnÃ©es JSON selon votre besoin
print(json_data)

# Vous pouvez Ã©galement enregistrer les donnÃ©es JSON dans un fichier
with open("post_infos.json", "w") as json_file:
    json.dump(post_infos, json_file, indent=4)
```
